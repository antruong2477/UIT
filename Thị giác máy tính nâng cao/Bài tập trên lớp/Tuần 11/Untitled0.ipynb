```python
import cv2
import os

def convert_video_to_images(video_path, output_folder, num_images):
    video = cv2.VideoCapture(video_path)
    total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))

    if total_frames == 0:
        print(f"Error: Could not determine the number of frames in {video_path}")
        return

    frame_interval = total_frames // num_images
    if frame_interval == 0:
        frame_interval = 1

    os.makedirs(output_folder, exist_ok=True)

    success, image = video.read()
    count = 0
    frame_count = 0

    while success and count < num_images:
      if frame_count % frame_interval == 0:
        cv2.imwrite(os.path.join(output_folder, f"frame_{count}.jpg"), image)
        count += 1

      success, image = video.read()
      frame_count += 1

video_folder = '/content/drive/MyDrive/CV nâng cao/Bài tập trên lớp/Tuần 11'
output_base_folder = "/content/drive/MyDrive/CV nâng cao/Bài tập trên lớp/Tuần 11/output_images"
num_images_per_video = 30

for filename in os.listdir(video_folder):
    if filename.endswith(".mp4"):
        video_path = os.path.join(video_folder, filename)
        output_folder = os.path.join(output_base_folder, filename[:-4])
        convert_video_to_images(video_path, output_folder, num_images_per_video)
        print(f"Converted {filename} to images in {output_folder}")
```
