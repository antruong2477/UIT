```python
import pandas as pd
```

# 01.sinhvien

```python
dir = "../../Dataset/UIT/01.sinhvien.xlsx"
df = pd.read_excel(dir)
```

```python
#Các cột trong df
df.columns
```

### Phân tích các biến rời rạc

```python
cac_cot_roi_rac = [' gioitinh', ' noisinh', ' lopsh', ' khoa', ' hedt', ' khoahoc', ' chuyennganh2', ' tinhtrang', ' diachi_tinhtp'] # chổ này nè
```

```python
df.dropna(axis=0, how='all', subset=None, inplace=True)
missing_values_count = df.isnull().sum()
print("Số mẫu: ", df.shape[0])
print("Số mẫu không hợp lệ:\n")
for index, value in missing_values_count.items():
    print(index, value)
roi_rac = []
for c in cac_cot_roi_rac:
    a = df[c].unique().shape[0]
    roi_rac.append(f" {c} {a}")
print("_________________________________________________")
print("Số giá trị đặc trưng (biến rời rạc)\n" ,roi_rac)
```

```python
for c in cac_cot_roi_rac:
    value_counts = df[c].value_counts()
    min_count = value_counts.min()
    least_frequent_values = value_counts[value_counts == min_count]
    max_count = value_counts.max()
    most_frequent_values = value_counts[value_counts == max_count]
    print(f"Các giá trị xuất hiện ít nhất là:{least_frequent_values}")
    print(f"Các giá trị xuất hiện nhiều nhất là:{most_frequent_values}")
    print("===============================")
```

### Phân tích các biến liên tục

```python
cot_lien_tuc = [' namsinh']
```

```python
for c in cot_lien_tuc:
    print(c)
    print("Mean",df[c].mean().round(4))
    print("Median",df[c].median().round(4))
    print("Variance",df[c].var().round(4))
    print("Standard Deviation",df[c].std().round(4))
    print("Min", df[c].min())
    print("Max", df[c].max())
```
