# **diem**

```python
cols = ['id', 'mssv', 'mamh', 'malop', 'sotc', 'namhoc', 'hocky', 'diem', 'trangthai', 'mamh_tt']
# for c in cols:
#   unique_values = raw_score_df[c].unique()
#   print(unique_values)
raw_score_df['mamh_tt'].unique()
```

```python
erro_val = [' NULL', ' NULL);', ' NULL)']
values_to_replace_count = raw_score_df['mamh_tt'].isin(erro_val).sum()

# In ra số lượng các giá trị sẽ thay thế
print(f"Số lượng giá trị cần thay thế (Số lượng lỗi): {values_to_replace_count}")
print(f"Tỷ lệ lỗi của cột: {values_to_replace_count / raw_score_df.shape[0] * 100:.2f}%")
print("Tỷ lệ lỗi của file dữ liệu:" , (values_to_replace_count / raw_score_df.shape[0] * 100)/len(cols),"%")

# Thay thế các giá trị trong erro_val bằng 'NULL'
raw_score_df['mamh_tt'] = raw_score_df['mamh_tt'].replace(erro_val, 'NULL')
```

```python
# Độ lệch chuẩn của cột diem
std_deviation = raw_score_df['diem'].std()

# Giá trị trung bình của cột diem
mean_value = raw_score_df['diem'].mean()

print(f"Độ lệch chuẩn của cột diem: {std_deviation}")
print(f"Giá trị trung bình của cột diem: {mean_value}")
```
