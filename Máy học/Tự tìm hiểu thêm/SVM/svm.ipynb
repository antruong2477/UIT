```python
import pandas as pd
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
iris = load_iris()
dir(iris)
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df['target'] = iris.target
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x]) # he he he he he he he he he he he he he he he he he he he
df
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'], marker='+',color='red')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'], marker='+',color='blue')
plt.show()
```

```python
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'], marker='+',color='red')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'], marker='+',color='blue')
plt.show()
```

```python
df
```

```python
X = df.drop(['target','flower_name'],axis=1)
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
model = SVC(C=10)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
df = pd.DataFrame(digits.data)
df['target'] = digits.target
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis=1), df.target, test_size=0.2)
```

## Using RBF kernel

```python
from sklearn.svm import SVC
rbf_model = SVC(kernel='rbf')
```

```python
rbf_model.fit(X_train, y_train)
```

```python
rbf_model.score(X_test,y_test)
```

## Using Linear kernel

```python
linear_model = SVC(kernel='linear')
linear_model.fit(X_train,y_train)
```

```python
linear_model.score(X_test,y_test)
```
